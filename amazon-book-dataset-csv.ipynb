```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#  --- Genre Popularity ---

# Bar chart: Total books per main genre

genre_counts = amazon_genres['Title'].value_counts().sort_values(ascending=False)
plt.figure(figsize=(12, 7))
plt.bar(genre_counts.index, genre_counts.values)
plt.xticks(rotation=90)
plt.xlabel('Genre')
plt.ylabel('Number of Books')
plt.title('Distribution of Books Across Genres')
plt.show()
```

```python
# Sub-genre proportions within top genres

top_genres = genre_counts.head(5).index  # Adjust the number of top genres as desired
for genre in top_genres:
    subgenres_in_genre = amazon_sub_genres[amazon_sub_genres['Main Genre'] == genre]
    plt.figure(figsize=(8, 5))
    plt.pie(subgenres_in_genre['No. of Books'], labels=subgenres_in_genre['Title'], autopct='%1.1f%%')
    plt.title(f"Sub-Genre Distribution for {genre}")
    plt.show()
```
