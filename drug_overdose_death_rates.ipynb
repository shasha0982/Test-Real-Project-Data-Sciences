```python
# @title Any opioid death rates (CDC WONDER)

from matplotlib import pyplot as plt
drug['Any opioid death rates (CDC WONDER)'].plot(kind='line', figsize=(8, 4), title='Any opioid death rates (CDC WONDER)')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
# @title Cocaine overdose death rates (CDC WONDER)

from matplotlib import pyplot as plt
drug['Cocaine overdose death rates (CDC WONDER)'].plot(kind='line', figsize=(8, 4), title='Cocaine overdose death rates (CDC WONDER)')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
# from 2000 year - 2005 years, the cocaine overdose death rates ( cdc wonder) have 3% increased into value '0 - 3', and from 2005 - 2020, the cocaine overdose death rates (cdc wonder) have increased +6% into value up to '6'.
```

```python
# @title Heroin overdose death rates (CDC WONDER)

from matplotlib import pyplot as plt
drug['Heroin overdose death rates (CDC WONDER)'].plot(kind='line', figsize=(8, 4), title='Heroin overdose death rates (CDC WONDER)')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
# from 2000 year - 2010 years, the of heroin overdose death rates are decreased into value '0 - 1'  and from 2011 - 2020,the heroin overdose death rates are increased into values '1 - 5' .
```

```python
# @title Synthetic opioids death rates (CDC WONDER)

from matplotlib import pyplot as plt
drug['Synthetic opioids death rates (CDC WONDER)'].plot(kind='line', figsize=(8, 4), title='Synthetic opioids death rates (CDC WONDER)')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
# from 2000 year - 2015 years, its have achieved least value are '0.0 - 2.5' and from 2016 - 2020, its have achieved most value are '2.6 - 17.5' synthetic opioids death rates
```
