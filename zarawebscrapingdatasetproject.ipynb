```python
from bs4 import BeautifulSoup
import requests
```

```python
# URL of the webpage we want to scrape

url = 'https://www.zara.com/us/en/'
```

```python
# Sending HTTP request to the specified URL and save the response from server in a response object called response

response = requests.get(url)
```

```python
# Creating BeautifulSoup object and specifying the parser library at the same time

soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Now, let's say we want all the URLs within the webpage. `a` tags are used for links in HTML. So, we should find all the `a` tags.

links = soup.find_all('a')
```

```python
# Here, I'm printing all the URLs we just scraped

for link in links:
    print(link.get('href'))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
zara_store['product_type'] = zara_store['name'].apply(lambda x: str(x).split(' ')[0].lower() if pd.notnull(x) else 'unknown')
```

```python
zara_store.head(80)
```

```python
# Now let's create a bar plot to visualize the number of items per product type
# For this purpose, we'll use the seaborn library

# First, let's calculate the count of each product type

product_counts = zara_store['product_type'].value_counts()
```

```python
# We will only display the top 10 product types for clarity

plt.figure(figsize=(10,6))
sns.barplot(x=product_counts.index, y=product_counts.values, alpha=0.8)
plt.title('Number of Items per Product Type')
plt.ylabel('Number of Items', fontsize=12)
plt.xlabel('Product Type', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# let's visualize the distribution of prices across all products using a histogram
```

```python
# Let's first check the data type of 'price'

zara_store['price'].dtype
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=zara_store, x="price", bins=30, log_scale=True, color='skyblue', edgecolor='black')
plt.title('Distribution of Prices')
plt.ylabel('Number of Items')
plt.xlabel('Price (Log Scale)')
plt.show()
```

```python
# using bar chart for visualization
```

```python
# Calculate the average price for each product type
average_prices = zara_store.groupby('product_type')['price'].mean().sort_values(ascending=False)

# Take the top 10 product types with highest average price
average_prices_top10 = average_prices[:10]

# Generate the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=average_prices_top10.index, y=average_prices_top10.values, palette='viridis')
plt.title('Average Price per Product Type', fontsize=15)
plt.xlabel('Product Type', fontsize=12)
plt.ylabel('Average Price', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```
