```python
# Data Analysis
```

```python
# Fill the missing data using the forward-fill method
climate_change_filled = climate_change.fillna(method='ffill')

# verify if there are still any missing values left in the data
climate_change_filled.isna().sum().sum()
```

```python
# time series line plot for the global average temperature change over these years
```

```python
import matplotlib.pyplot as plt

# Assuming climate_change_filled and year_data are defined elsewhere

# Extract year data and transpose the dataframe
year_data = climate_change_filled.loc[:, 'F1961':'F2022'].T
year_data.columns = climate_change_filled['Country']

# Calculate the global average temperature change for each year
global_avg_temp_change = year_data.mean(axis=1)

# Plot a line graph
plt.figure(figsize=(10, 6))
plt.plot(global_avg_temp_change)

# Adjust x-axis ticks for readability
plt.xticks(rotation=45)  # Rotate x-axis labels by 45 degrees
plt.title('Global Average Temperature Change (1961 - 2022)')
plt.xlabel('Year')
plt.ylabel('Temperature Change (°C)')  # Adding ylabel for clarity
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# the average global temperature has been rising over the past six decades
```

```python
# plot the temperature change over the years
```

```python
import matplotlib.pyplot as plt

# Assuming year_data is defined elsewhere

# Plotting temperature change for selected countries
plt.figure(figsize=(12, 8))

# Plot for United States
plt.plot(year_data.index, year_data['United States'], label='United States')

# Plot for China
plt.plot(year_data.index, year_data['China, P.R.: Mainland'], label='China')

# Plot for India
plt.plot(year_data.index, year_data['India'], label='India')

# Adjust x-axis ticks for readability
plt.xticks(rotation=45)  # Rotate x-axis labels by 45 degrees

plt.title('Temperature Change (1961 - 2022) for Selected Countries')
plt.xlabel('Year')
plt.ylabel('Temperature Change (in Degree Celsius)')
plt.legend()
plt.grid()
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# The line plot illustrates the temperature change from 1961 to 2022 for the selected countries: United States, China, and India.
# These countries shows a general upward trend
```

```python
# plot bar graphs for the ten countries with the highest average temperature changes in the last
```

```python
# For clustering, we need to define the boundaries.
# Let's get the global average temperature change and its standard deviation to define our clusters

mean_temp_change = year_data.mean().mean()
std_temp_change = year_data.mean().std()

# Defining the boundaries of clusters
low_boundary = mean_temp_change - std_temp_change
high_boundary = mean_temp_change + std_temp_change

# Categorizing countries based on their average temperature change over years
def cluster_temp_change(temp_change):
  if temp_change < low_boundary:
    return 'Low Temperature Change'
  elif temp_change > high_boundary:
    return 'High Temperature Change'
  else:
    return 'Medium Temperature Change'

# Apply the function to the entire dataframe
countries_cluster = year_data.mean().apply(cluster_temp_change)

# Look at the number of countries in each cluster
countries_cluster.value_counts()
```

```python
# This simple categorization helps us understand which countries experienced high, medium, and low increases in temperatures from 1961 to 2022.
```

```python
# plot a bar graph for the ten countries with the highest average temperature changes from 2013 to 2022
```

```python
# Calculate average temperature change in the last decade for each country
last_decade_temp_change = year_data.loc['F2013':'F2022'].mean()

# Get top 10 countries with the highest average temperature change in the last decade
top_10_countries = last_decade_temp_change.nlargest(10)

# Plotting
plt.figure(figsize=(12, 7))
top_10_countries.sort_values().plot(kind='barh', color='skyblue', edgecolor='black')
plt.title('Top 10 Countries with Highest Average Temperature Change in Last Decade (2013 - 2022)')
plt.xlabel('Temperature Change (in Degree Celsius)')
plt.ylabel('Country')
plt.grid(axis='x')
plt.show()
```

```python
# It's clear that the countries, namely, Chad, Western Sahara, Mauritania, Mali, Syria, Guinea-Bissau, Gabon, Senegal, and Paraguay are experiencing significant temperature changes, indicating the impact of climate change in these regions.
```
