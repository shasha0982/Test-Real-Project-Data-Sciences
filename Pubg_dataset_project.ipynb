```python
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def get_freq_words(text, n):
    words = text.split()
    word_counts = Counter(words)
    return word_counts.most_common(n)

def create_wordcloud(text):
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

# Creating a single string for df_1 content
pubg_2_8_text = ' '.join(pubg_2_8['content'])

# Creating a single string for df_2 content
pubg_2_9_text = ' '.join(pubg_2_9['content'])

# Getting top 10 frequent words
pubg_2_8_freq_words = get_freq_words(pubg_2_8_text, 10)
pubg_2_9_freq_words = get_freq_words(pubg_2_9_text, 10)

# Creating word cloud for df_1
print("Word Cloud for df_1")
create_wordcloud(pubg_2_8_text)

# Creating word cloud for df_2
print("Word Cloud for df_2")
create_wordcloud(pubg_2_9_text)

# print top 10 frequent words

print("Top 10 frequent words for pubg_2_8:", pubg_2_8_freq_words)
print("Top 10 frequent words for pubg_2_9:", pubg_2_9_freq_words)
```
