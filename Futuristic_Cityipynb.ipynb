```python
# Analyzing relationships between numeric variables 'Speed', 'Energy Consumption', 'Traffic Density',

# Running correlation matrix

correlation = city[['Speed', 'Energy Consumption', 'Traffic Density']].corr()

correlation
```

```python
# The correlation matrix reveals:

# 1. There's a strong positive correlation between 'Speed' and 'Energy Consumption' (0.84). As the speed of the vehicle increases, the energy consumption also tends to increase.
# 2. 'Traffic Density' has very weak negative correlation with 'Speed' and very weak positive correlation with 'Energy Consumption'. The correlations are so small they're practically negligible. While it's likely that these parameters could influence each other, from the current dataset we can't observe this behavior.
```

```python
# Count plots for Weather and Economic Condition

# Creating subplots
fig, axes = plt.subplots(2, 1, figsize=(10,12))

# Weather

sns.countplot(ax=axes[0], data=city, x='Weather', order = city['Weather'].value_counts().index)
axes[0].set_title('Distribution of Traffic Across Different Weather Conditions')
axes[0].set_xlabel('')
axes[0].tick_params(axis='x', rotation=50)

# Economic Condition
sns.countplot(ax=axes[1], data=city, x='Economic Condition', order = city['Economic Condition'].value_counts().index)
axes[1].set_title('Distribution of Traffic Across Different Economic Conditions')
axes[1].set_xlabel('')


plt.tight_layout()
plt.show()
```

```python
# The bar plots reveal the following insights:

# 1. Distribution of Traffic Across Different Weather Conditions : The data shows that most traffic occurs during 'Clear' weather conditions.
# This is followed by 'Cloudy' and 'Rainy' conditions. The least amount of traffic is seen during an 'Electromagnetic Storm'.

# 2. Distribution of Traffic Across Different Economic Conditions : The majority of traffic occurs during 'Stable' and 'Booming' economic conditions,
# with 'Stable' conditions leading by a small margin. The least amount of traffic is observed during 'Recession' periods.
```

```python
# # Box plots for 'Speed' and 'Energy Consumption' vs 'Is Peak Hour'
# Box plots for 'Traffic Density' and 'Speed' vs 'Random Event Occurred'
```

```python
# Creating subplots
fig, axes = plt.subplots(2, 2, figsize=(18,10))

# Speed vs Is Peak Hour
sns.boxplot(ax=axes[0, 0], data=city, x='Is Peak Hour', y='Speed')
axes[0, 0].set_title('Speed vs Is Peak Hour')
axes[0, 0].set_xticklabels(['No Peak Hour', 'Peak Hour'])
axes[0, 0].set_xlabel('')

# Energy Consumption vs Is Peak Hour
sns.boxplot(ax=axes[0, 1], data=city, x='Is Peak Hour', y='Energy Consumption')
axes[0, 1].set_title('Energy Consumption vs Is Peak Hour')
axes[0, 1].set_xticklabels(['No Peak Hour', 'Peak Hour'])
axes[0, 1].set_xlabel('')

# Traffic Density vs Random Event Occurred
sns.boxplot(ax=axes[1, 0], data=city, x='Random Event Occurred', y='Traffic Density')
axes[1, 0].set_title('Traffic Density vs Random Event Occurred')
axes[1, 0].set_xticklabels(['No Event', 'Event Occurred'])
axes[1, 0].set_xlabel('')

# Speed vs Random Event Occurred
sns.boxplot(ax=axes[1, 1], data=city, x='Random Event Occurred', y='Speed')
axes[1, 1].set_title('Speed vs Random Event Occurred')
axes[1, 1].set_xticklabels(['No Event', 'Event Occurred'])
axes[1, 1].set_xlabel('')

plt.tight_layout()
plt.show()
```

```python
# Speed vs Is Peak Hour: There's a noticeable drop in median speed during peak hours, which is an expected pattern in urban traffic.
# The interquartile range (IQR) for speed is wider in non-peak hours, showing more variation in the speed of vehicles.

# Energy Consumption vs Is Peak Hour : Energy consumption appears slightly higher during peak hours,
# which could be a result of more stop-and-go traffic (requiring more energy to accelerate) or other factors.

# Traffic Density vs Random Event Occurred: When a random event occurs, the median traffic density shows a slight increase.
# This could be due to disruption in normal traffic flow caused by the event.

# Speed vs Random Event Occurred: Speed tends to decrease slightly when a random event occurs, possibly indicating disruptions slowing down traffic.
```
